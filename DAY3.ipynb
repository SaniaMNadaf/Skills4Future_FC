```python
#feature and target selection
x=df[['carbon_emissions','energy_output','renewability_index','cost_efficiency']]
y=df['adoption']
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.8,random_state=42)
```

```python
#instantiate
model=DecisionTreeClassifier(max_depth=3,random_state=42)
```

```python
#train model
model.fit(xtr,ytr)
```

```python
#predict
y_pred=model.predict(xte)
y_pred
```

```python
#evaluate 
acc=accuracy_score(yte,y_pred)
print(f'The accuracy score of DTC is{acc*100:.2f}%')
```

```python
cm=confusion_matrix(yte,y_pred)
sns.heatmap(cm,annot=True,cmap='Blues')
xticklabels=['Non Adoption','Adoption'],
yticklables=['Non Adoption','Adoption']
plt.title('Heatmap of Confusion Matrix')
plt.xlabel('Predicted Values')
plt.ylabel('Actual values')
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,10))
plot_tree(model,filled=True,feature_names=x.columns,class_names=['Non Adoption','Adoption'])
plt.show()
```
