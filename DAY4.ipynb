```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
#load data set
df=pd.read_csv(r"C:\Users\Ahmed\OneDrive\Documents\agricultural sustainability.csv")
```

```python
#featur and target selection
x=df[['soil_health', 'crop_yield', 'water_usage', 'carbon_footprint', 'fertilizer_use']] 
y=df['sustainability']
```

```python
#spliting train and test
xtr,xte,ytr,yte=train_test_split(x,y,train_size=0.8,random_state=11)
```

```python
xtr.shape
```

```python
xte.shape
```

```python
#instantiate the modelRF
model=RandomForestClassifier(n_estimators=100,random_state=100)
```

```python
#train the model
model.fit(xte,yte)
```

```python
#prediction
y_pred=model.predict(xte)
y_pred
```

```python
len(y_pred)
```

```python
#evaluate
acc=accuracy_score(yte,y_pred)
print(f'The accuracy score for RFC is {acc*100:.2f}%')
```

```python
#confusion matrix
cm=confusion_matrix(yte,y_pred)
sns.heatmap(cm,annot=True,cmap='Blues')
xticklabel=['Unsustainable','Sustainable']
yticklabel=['Unsustainable','Sustainable']
plt.title('Heatmap of confusion Matrix')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
#classification report
cr=classification_report(yte,y_pred)
print(cr)
```

```python
#save model
import joblib
joblib.dump(model,'RF_Model.pkl')
```
